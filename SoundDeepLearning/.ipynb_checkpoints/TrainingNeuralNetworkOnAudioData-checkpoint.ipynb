```python
class SoundData(Dataset):
    
    def __init__(self, annotations):
        self.DESTINATION = '/home/grey/Music/AUDIO_RECOG/DESTINATION/'
        self.annotations = annotations 
        self.device =  device 
        
        
    def __len__(self):
        
        return len(self.annotations)
    

    
    def __getitem__(self,index):
            audio_filename =  self.DESTINATION+self.annotations['AudioName'].iloc[index]
            label =  self.annotations['Onehot'].iloc[index]
            
            
            n_fft = 1024  # Number of FFT points
            hop_length = 512  # Hop length in samples
            n_mels = 128  # Number of Mel bands
            
            signal,sr =  librosa.load(audio_filename)
            n_mfcc = 80
#             mel_spectrogram = librosa.feature.melspectrogram(y=signal, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
#             mel_spectrogram_db = librosa.amplitude_to_db(mel_spectrogram, ref=np.max)
            mfccs = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc)
    
#             final_output = mel_spectrogram_db.reshape(1,128,216)
            mfccs =  mfccs.reshape(1,n_mfcc,216)
            return mfccs,label
 
```

```python
class YourModel(nn.Module):
    def __init__(self, num_classes):
        super(YourModel, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1).to(device)
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1).to(device)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2).to(device)
        self.conv3 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1).to(device)
        self.flatten = nn.Flatten().to(device)
        self.fc1 = nn.LazyLinear( 128).to(device)
        self.fc2 = nn.Linear(128, num_classes).to(device)
        
        self.dropout =  nn.Dropout(p=0.4)
        
        

    
    def forward(self, x):
        x = self.conv1(x)
#         x= self.dropout(x)
        x = nn.ReLU()(x)
        x = self.pool(x)
        
        
        x = self.conv2(x)
        x = nn.ReLU()(x)
        x = self.pool(x)
        
        
        x =  self.conv3(x)
#         x= self.dropout(x)
        x =  nn.ReLU()(x)
        x = self.pool(x)
        
        x = self.flatten(x)
        x = self.fc1(x)
        x =  self.dropout(x)
        x = nn.ReLU()(x)
        x = self.fc2(x)
        x =  self.dropout(x)

        
        return x
```

```python
train_size = int(0.8 * len(mfcc_data))
val_size = len(mfcc_data) - train_size

train_dataset, val_dataset = random_split(mfcc_data, [train_size, val_size])
```

```python
batch_size = 128

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
```

```python
class ConvNetwork(nn.Module):
    
    
    def __init__(self):
        super().__init__()
    
    
        self.conv1 =  nn.Sequential(
    
            nn.Conv2d(
                in_channels=1,
                out_channels=16,
                kernel_size=3,
                stride=1,
                padding=2                     
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
            
    )
    
        self.conv2 = nn.Sequential(
    
            nn.Conv2d(
                in_channels=16,
                out_channels=32,
                kernel_size=3,
                stride=1,
                padding=2                     
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
            
    )
    
        self.conv3 = nn.Sequential(
    
            nn.Conv2d(
                in_channels=32,
                out_channels=64,
                kernel_size=3,
                stride=1,
                padding=2                     
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),
            nn.Dropout(p=.4)
            
    )
    
        self.conv4 =nn.Sequential(
    
            nn.Conv2d(
                in_channels=64,
                out_channels=128,
                kernel_size=3,
                stride=1,
                padding=2                     
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
            
    )
        self.conv5 =nn.Sequential(
    
            nn.Conv2d(
                in_channels=128,
                out_channels=256,
                kernel_size=3,
                stride=1,
                padding=2                     
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
            
    )
        self.conv6 =nn.Sequential(
    
            nn.Conv2d(
                in_channels=256,
                out_channels=512,
                kernel_size=3,
                stride=1,
                padding=2                     
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
            
    )
    
        self.flatten =  nn.Flatten()
        
#         self.linear =  nn.Linear(512 *2*5,7)
        self.linear =  nn.LazyLinear(7)
        self.softmax =  nn.Softmax(1)
        
    def forward(self,input_data):
        x =  self.conv1(input_data)
        x =  self.conv2(x)
        x =  self.conv3(x)
        x = self.conv4(x)
        x =  self.conv5(x)
        x =  self.conv6(x)
        
        x =  self.flatten(x)
        
        loggits  = self.linear(x)
        
        predictions =  self.softmax(loggits)
        
        return predictions
    
    
```

```python
def prepareAudio(audio):
    
    n_fft = 1024  # Number of FFT points
    hop_length = 512  # Hop length in samples
    n_mels = 128  # Number of Mel bands
            
    signal,sr =  librosa.load(audio)
            
    mel_spectrogram = librosa.feature.melspectrogram(y=signal, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
    mel_spectrogram_db = librosa.amplitude_to_db(mel_spectrogram, ref=np.max)
            
    final_output = mel_spectrogram_db.reshape(1,128,216)
    return final_output
```

```python
def resample(audio):
    desired_sample_rate  = 22050
    # Check the number of channels

    audio_segment = audio.set_frame_rate(desired_sample_rate)

# Convert stereo audio to mono
    if audio_segment.channels == 2:
        audio_segment = audio_segment.set_channels(1)

# Convert AudioSegment to a NumPy array
    audio_data = audio_segment.get_array_of_samples()

# Perform noise reduction using noisereduce
    reduced_audio_data = nr.reduce_noise(y=audio_data, sr=desired_sample_rate)

# Convert the reduced audio data back to an AudioSegment
    reduced_audio_segment = AudioSegment(
        reduced_audio_data.tobytes(),
        frame_rate=desired_sample_rate,
        sample_width=audio_segment.sample_width,
        channels=1  # The reduced audio is mono
    )
    
    return reduced_audio_segment
```
